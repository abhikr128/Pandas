### Multiple conditions
- Use &, |, ~(Bitwise operators) not (and, or, not)
- Enclose each conditional expressions in parethesis when using comparision operators

```python
cols = ['name', 'survived', 'pclass', 'age', 'sex', 'fare']
titanic = pd.read_excel('titanic3.xls', usecols=cols, index_col='name')
titanic
```

```python
# Selecting all the females that were in 1st class who also survived

female_mask = titanic['sex'] == 'female'
class_mask = titanic['pclass'] == 1
survived_mask = titanic['survived'] == 1
titanic[female_mask & class_mask & survived_mask]
```

```python
# Selecting the richest kid on the plane

rich_kids_mask = (titanic['age'] < 21) & (titanic['fare'] > 200)
titanic[rich_kids_mask]
```

```python
rich = titanic[rich_kids_mask]
rich.nlargest(n=3, columns=['age', 'fare'])
```

```python
# Selecting  all the people that were under 21 or females that survived_mask

age_mask = titanic['age'] <= 21
female_mask = titanic['sex'] == 'female'
survived_mask = titanic['survived'] == 1
titanic[(age_mask | female_mask) & survived_mask]
```

```python
titanic.loc[(age_mask | female_mask) & survived_mask, 'age']    # This works too
```
