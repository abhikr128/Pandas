### Pie Chart

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
richest = pd.read_csv('TopRichestInWorld.csv')
richest
```

```python
industries = richest['Industry'].value_counts()
industries
```

```python
industries.index
```

#### Creating our first pie chart

```python
plt.pie(x=industries.values, labels=industries.index)
plt.show()
```

#### Let's display the top 10 most popular industries with percentage

```python
labels = industries.index[:10]
values = industries.values[:10]

plt.pie(x=values, labels=labels, autopct='%.2f%%')
plt.show()
```

#### We can specify our own colors too

```python
plt.Figure()
plt.pie(x=values, labels=labels, colors=('skyblue', 'orange', 'green'))
plt.show()
# If you don't specify all the colors, it will loop through them
```

### Style

```python
pokemons = pd.read_csv('Pokemon.csv', index_col='#')
pokemons
```

```python
poke_groups = pokemons.groupby(by='Type 1')
max_attack = poke_groups['Attack'].max()
max_attack
```

```python
max_attack.plot(kind='barh', color=('skyblue', 'crimson'))
```

#### Let's style our plot using a style instead

```python
plt.style.available
```

```python
plt.style.use('grayscale')
max_attack.plot(kind='barh')
```

```python
plt.style.use('seaborn-v0_8-colorblind')
max_attack.plot(kind='barh')
```
