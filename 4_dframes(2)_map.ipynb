### map() and applymap()

```python
import pandas as pd
```

```python
pokemons = pd.read_csv('Pokemon.csv', index_col='#')
pokemons
```

### map() - map a series

```python
pokemons['Legendary'] = pokemons['Legendary'].map({True:'Yes', False:'No'})
pokemons
```

```python
pokemons['Generation'].map({1:'first', 2:'second'}).value_counts()     # Values that are specified will be set to NaN
    
```

### applymap() - applies a function throughout a dataframe

```python
pokemons.applymap(type)
```

```python
def uppercase(var: str) -> str:
    return var.upper()
```

```python
pokemons = pokemons.applymap(str)
pokemons.applymap(uppercase)
pokemons.info()
```

```python
pokemons.applymap(len)
```
