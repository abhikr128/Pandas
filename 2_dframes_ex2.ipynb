## nlargest() and nsmallest() - to sort data in our dataframes

````python
import pandas as pd
````

````python
cols = ['Series_Title', 'IMDB_Rating', 'No_of_Votes']
movies = pd.read_csv('imdb_top_1000.csv', usecols=cols)
movies
````

### .nlargest() 
- This method is eqvt to ```df.sort_values(columns, ascending=False).head(n)```, but more performant

````python
movies.nlargest(n=10, columns=['No_of_Votes'])
````

````python
movies.nlargest(n=10, columns=['IMDB_Rating', 'No_of_Votes'])   # Priority matters in the list how you put
````

````python
movies.nsmallest(n=10, columns=['No_of_Votes'])
````

````python
movies.nsmallest(n=10, columns=['IMDB_Rating', 'No_of_Votes'])   # Priority matters in the list how you put
````

- keep='first' means it will keep the first duplicates
- keep='last' means it will keep the last duplicates
- keep='all' means it will keep the all duplicates

````python
movies.nlargest(n=10, columns=['IMDB_Rating'], keep='first')
````

### .sort_values() 
- You can still use sort_values(). It's less efficient and less optimized than using nlargest or nsmallest, which means it can actually take longer to perform an operation.

````python
movies.sort_values(by=['IMDB_Rating', 'No_of_Votes'], ascending=False).head(10)
````
