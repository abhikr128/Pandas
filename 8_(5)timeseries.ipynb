### loc[] and iloc[] with DatetimeIndex

```python
import pandas as pd
```

```python
cols = ['date', 'high', 'low', 'open', 'close']
apple_stock = pd.read_csv('AAPL.csv', usecols=cols, index_col='date', parse_dates=['date'])
apple_stock
```

#### Using iloc remains same with datetime objects

```python
apple_stock.iloc[0]
```

```python
apple_stock.iloc[0:5]
```

```python
apple_stock.iloc[:5]
```

#### It changes when we use loc though. We can select rows without selecting everything

```python
apple_stock.loc['2015-05-27']
```

```python
apple_stock.loc['2015.07.27':'2015-07-31']
```

```python
apple_stock.loc['2015-07']
```
