### apply() applies functionto a series

```python
import pandas as pd
```

```python
pokemons = pd.read_csv('Pokemon.csv', index_col='#')
pokemons
```

```python
pokemons['name_length'] = pokemons['Name'].apply(len)    # Apply the len() function to each element
pokemons
```

```python
del pokemons['name_length']    # Delete the name_length column
```

```python
def strength(attack: int) -> str:
    if attack > 150:
       return 'Strong'
    elif attack > 100:
       return 'Average'
    else:
       return 'Weak'
```

```python
pokemons['Strength'] = pokemons['Attack'].apply(strength)
pokemons
```

```python
pokemons.value_counts('Strength')
```

```python
del pokemons['Strength']
```

### .apply() can also use it with multiple columns

```python
def score(row):
    attack: int = row['Attack']
    defense: int = row['Defense']
    speed: int = row['Speed']

    return sum([attack, defense, speed])    
```

```python
pokemons['Score'] = pokemons.apply(score, axis=1)
pokemons
```

### we can also use lamda()

```python
short_name = pokemons['Name'].apply(lambda x: x[:5])
pokemons.insert(1, 'Short', short_name)
pokemons
```
