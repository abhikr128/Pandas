### Rename()

````python
import pandas as pd
````

````python
cols = ['name', 'survived', 'pclass', 'fare', 'age', 'sex']
titanic = pd.read_excel('titanic3.xls', index_col='name', usecols=cols)
titanic
````

#### one way to rename columns is to pass in a list of names of the same length

````python
titanic.columns = ['Class', 'Survived', 'Gender', 'Age', 'Ticket Price']
titanic
````

#### A more convinient way is to use rename()

````python
titanic.rename(columns={'Ticket Price': 'Fare'}, inplace=True)
titanic
````

#### By default, all the operations are performed on the zero axis. You can still choose to specify which axis you want via ```axis``` keyword, or you can also directly use ```index=``` and ```column=```

````python
titanic.rename({'Fare':'Cost', 'Class':'Group'}, axis=1)
````

````python
titanic.rename({'Allen, Miss. Elisabeth Walton':'First Passenger'}, axis=0)      # axis=0 is default
````

#### specifying an index name and column name

````python
titanic.index.name = 'Passengers'
titanic.columns.name = 'Columns'
titanic
````
