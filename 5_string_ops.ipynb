### String operations
- Note : Strings are object but objects are not string

```python
import pandas as pd
```

```python
pokemons = pd.read_csv('Pokemon.csv', index_col='#')
pokemons
```

```python
# pokemons['Name'].upper()    # Does not work as dataframe treat it as object not string.
```

```python
# pokemons['Name'].astype(str).upper()   # Does not work either
```

```python
pokemons['Name'].str.upper()       # It works
```

##### Everytime you want to use string method, you need to use .str first

```python
# pokemons['Attack'].str.upper()      # It doesn't work with non-string values
```

```python
pokemons['Name'].str.lower()         # Lower string
```

```python
pokemons['Name'].str[::-1]      # Reverse string
```

##### Can also be used for filtering

```python
x_filter = pokemons['Name'].str.contains('q', case=False)      # Whether to ignore case or not
pokemons[x_filter]
```

#### Some other string opearations
- .startswith()
- .endswith()
- .replace()

```python
name_filter = pokemons['Name'].str.startswith('Do')      # It's case sensitive
pokemons[name_filter]
```

```python
name_filter = pokemons['Name'].str.endswith('chu')      # It's case sensitive
pokemons[name_filter]
```

```python
pokemons['Type 1'].str.replace('Grass', 'Vegan')     
```

```python
pokemons['Type 1'] = pokemons['Type 1'].str.replace('Grass', 'Vegan')
pokemons[pokemons['Type 1'] == 'Vegan']    # Display all the vegan pokemons
```

#### String ops on index and columns

```python
pokemons = pd.read_csv('Pokemon.csv', index_col='Name')
pokemons
```

```python
# Using .str on columns
pokemons.columns.str.upper()
```

```python
pokemons.columns = pokemons.columns.str.upper()    # We can use this to rename our columns
pokemons
```

```python
# Using .str on index
pokemons.index.str.upper()
```

```python
pokemons.index.str[:5]      # We can use this to rename our entire index
```

```python
pokemons.index = pokemons.index.str[:5]
pokemons
```
