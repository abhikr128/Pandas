### Dataset 1

```python
import pandas as pd
```

```python
cols = ['Series_Title', 'Genre', 'IMDB_Rating', 'Gross', 'Released_Year', 'No_of_Votes']
movies = pd.read_csv('imdb_top_1000.csv', usecols=cols)
movies
```

### info()

```python
movies.info()
```

```python
movies.describe().round()      # It only gives data regarding numerical values not objects so, we didn't get the result for Gross
```

### Converting Gross data which is object data type into numeric data type

```python
movies['Gross']=movies['Gross'].str.replace(',', '')
movies
```

```python
# Covert data into integer so we can use it as numeric data

movies['Gross'].fillna(0, inplace=True)          # fillna() replaces all NaN values to 0 where the data is NaN
movies['Gross']=movies['Gross'].astype('int64')  # .astype will convert into numerical data as int64
movies.describe().round()
```
