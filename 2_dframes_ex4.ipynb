### Adding columns to a dataframe

```python
import pandas as pd
```

```python
movies = pd.read_csv('imdb_top_1000.csv', usecols=['Series_Title', 'IMDB_Rating', 'No_of_Votes'])
movies
```

```python
movies['IMDB_Rating']
```

```python
movies[['IMDB_Rating', 'No_of_Votes']]
```

```python
# movies['Watched']      # Will give error as it doesn't exist
```

```python
movies['Watched'] = True       # Columns get inserted at last location by assigning the bool value
movies
```

```python
# Insert a column at specified location
movies.insert(loc=2, column='My_Rating', value=None)
movies
```
