### Hierarchical Indexing( Multi-Indexing)

```python
import pandas as pd
```

```python
pokemons = pd.read_csv('Pokemon.csv')        
pokemons
```

#### In Pandas, you can create multi index by passing in a slist of column names into set_index()

```python
pokemons.set_index(['Name', 'Type 1'], inplace=True)
pokemons
```

```python
pokemons.index        # Returns each index as a tuple
```

#### Use get_level_values() to specify which index you want to get back

```python
pokemons.index.get_level_values(0)
```

```python
# You can also swap the labels using swap_label
pokemons = pokemons.swaplevel()
pokemons
```

```python
pokemons.sort_values('Type 1')
```

### Multi-Indexing(Continued)

```python
pokemons = pd.read_csv('Pokemon.csv', index_col=['Type 1', 'Name'])
pokemons
```

##### Selecting rows

```python
pokemons.loc['Grass']        # Now, we can select all the Grass Pokemons
```

```python
pokemons.loc['Grass', 'Attack']      # Selecting a column
```

```python
pokemons.loc[('Grass', 'Bulbasaur')]       # You can also get more specific using tuple
```

```python
pokemons.loc[('Grass', 'Bulbasaur'), 'Attack']     # Selecting a column
```

```python
pokemons.loc[('Grass', 'Bulbasaur'), ['Attack', 'Defense']]  # Selecting multiple columns
```

#### Using mutiple indices and returning all the rows require ':' at the end

```python
pokemons.loc[(['Grass', 'Fire'], ['Bulbasaur', 'Litleo']), :]     # Using mutilple indices
```

#### Ignoring certain indices

```python
pokemons.loc[:, 'Bulbasaur', 'Legendary']    # You can use : to select all
# But you can't select a column using the syntax
```

```python
pokemons.loc[(:, 'Bulbasaur'), 'Legendary']  # Doesn't work
```

```python
pokemons.loc[(slice(None), 'Bulbasaur'), 'Legendary']    # One solution
```

```python
pokemons.loc[pd.IndexSlice[:, 'Bulbasaur'], 'Legendary']    # Another solution
```
