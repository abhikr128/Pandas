#### Line plots

```python
cols = ['date', 'high', 'low', 'open', 'close']
apple_stock = pd.read_csv('AAPL.csv', usecols=cols, index_col='date', parse_dates=['date'])
apple_stock.plot()
```

```python
apple_stock.loc['2016-05'].plot()
```

```python
apple_stock.loc['2016-05'].plot(y=['close', 'open'])
```

```python
apple_stock.plot(subplots=True, figsize=(12, 8))
```

```python
apple_stock.plot(title='Apple Stock', xlabel='Years', ylabel='Price', grid='True', figsize=(10, 8))
```

```python
apple_stock.loc['2016-05'].plot(y='close', c='g', linewidth=2, linestyle='-')     # use linestyle=':' for dotted line
```
