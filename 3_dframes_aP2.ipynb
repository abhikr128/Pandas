### ```loc[]``` allows us to select rows
- We can use label names and if the index is numeral, we can use numbers.

### ```iloc[]``` allows us to use to index position to select rows
- we can use numbers to access elements via their index position.

````python
# Selecting a row via index
titanic.loc['Abbott, Master. Eugene Joseph']
````

````python
# Selecting multiple row by index
# We can also select multiple rows by slicing
titanic.loc[['Abelseth, Miss. Karen Marie', 'van Billiard, Mr. Austin Blyler']]
````

````python
# If there are duplicates, it will returns them all
titanic.loc['Kelly, Mr. James']
````

````python
# Returns a value of the passenger for a specified column
titanic.loc['Abelseth, Miss. Karen Marie', 'fare']
````

````python
# Returns a series of the passengers with the specified columns
titanic.loc['Abbott, Mr. Rossmore Edward', ['fare', 'sex', 'cabin']]
````

````python
passengers = ['Abbing, Mr. Anthony', 'Abelseth, Miss. Karen Marie', 'Abbott, Mr. Rossmore Edward', 'Zabour, Miss. Thamine']
titanic.loc[passengers, 'fare']
````

````python
passengers = ['Abbing, Mr. Anthony', 'Abelseth, Miss. Karen Marie', 'Abbott, Mr. Rossmore Edward', 'Zabour, Miss. Thamine']
titanic.loc[passengers, ['fare', 'pclass', 'survived']]
````

````python
titanic = pd.read_excel('titanic3.xls', usecols=['name', 'sex', 'age', 'pclass', 'fare'])
titanic
````

````python
titanic.iloc[0]    # Select the row at the index of 0
````

````python
titanic.iloc[0:6]   # Can be used as python slicing syntax
````

````python
titanic.iloc[0:6:2]   # Can be used as python slicing syntax with a step of any numbers
````

````python
titanic.iloc[[100, 200, 300]]   # Selecting specific indices
````

````python
titanic.iloc[-3:]   # Selecting last 3 rows by negative index
````

### iloc[] continued...

````python
titanic = pd.read_excel('titanic3.xls', usecols=['name', 'sex', 'age', 'pclass', 'fare'])
titanic.head(10)
````

````python
titanic.iloc[5, 3]     # index position, column position
````

````python
titanic.iloc[[4, 5, 6], 3]  # Selecting age for passengers 4, 5 and 6
````

````python
titanic.iloc[5, [0, 1, 2]]     # Selecting col [0, 1, 2] for  passenger 5
````

````python
titanic.iloc[[4, 5, 6], [0, 1, 2]]   # Selecting col [0, 1, 2] for  passengers 4, 5 and 6
````

````python
titanic.iloc[:, [0, 1, 2]]      # Selecting col [0, 1, 2] for all passengers
````

### ```loc[]``` and ```iloc``` are good options, but using ```loc[]``` can definitely help with being more explicit since we refers to the labels rather than the index positions, making it easier to understand what we're doing.
