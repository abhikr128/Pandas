### .replace() 
- The benifit os using replace is that it only replaces the values that you specify(unlike when we use map)
- We can also use it on entire dataframe

```python
import pandas as pd
```

```python
pokemons = pd.read_csv('Pokemon.csv', index_col='#')
pokemons
```

```python
pokemons['Legendary'].replace({True:"Yes", False:"No"}, inplace=True)
pokemons
```

```python
pokemons['Generation'].replace({1:'First'})
pokemons
```

```python
# We can also use it on entire dataframe
pokemons.replace({'Grass':'!!', 'First':'!!!', 'Yes':'!!!!', 'hhjdbiw':'!!!'})
```

### .where() 

```python
pokemons = pd.read_csv('Pokemon.csv', index_col='#')
pokemons
```

#### We already know how to use a condition to select rows

```python
second_gen = pokemons['Generation'] == 2
pokemons[second_gen]
```

#### we can also use .where(), which instead will return NaN if the conditon is false

```python
pokemons.where(pokemons['Generation']==2)   # We can use dropna() to see what we have filtered
```

```python
pokemons = pokemons.where(pokemons['Type 1'] == 'Grass', other = None)
pokemons.dropna()
```
