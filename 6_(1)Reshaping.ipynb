### transpose()

```python
exams.transpose()     # Invert the columns and indices
```

```python
exams.T         # Another way to transpose
```

```python
transposed = exams.transpose()
transposed.loc['gender']          # All the columns are now indices and indices are now columns
```

##### Note: When you transpose(), your data type might change

```python
exams.info()         # Original
```

```python
t_exams = exams.transpose()
t_exams.info()                       # They all are now objects
```

```python
t2_exams = t_exams.transpose()
t2_exams                            # The transposed get transposed again and becomes original again but data types will not be changed to original
```

```python
t2_exams.info()      # Even it get changed to original, all the data types are objects.
```
