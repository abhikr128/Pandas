### melt() used to unpivot a dataframe from wide to long format
- ```id_vars``` are what we want to preserve
- ```values_vars``` are what we want to 'melt' into variables


````python
import pandas as pd
````

````python
medals = pd.read_csv('olympics.csv', sep=';')
medals
````

````python
medals.melt(id_vars='Country', value_vars=['2021', '2022', '2023', '2024'])
````

````python
medals                         # Note: melt() doesn't perform an inplace operation.
````

#### Providing names for vars and values

````python
medals.melt(id_vars='Country', var_name='Year', value_name='Medals', value_vars=['2022', '2023'])
````
