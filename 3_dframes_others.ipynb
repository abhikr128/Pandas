### between()

```python
import pandas as pd
```

```python
cols = ['name', 'survived', 'pclass', 'age', 'fare', 'sex']
titanic = pd.read_excel('titanic3.xls', usecols=cols, index_col='name')
titanic
```

```python
upper_age_mask = titanic['age'] <= 25
lower_age_mask = titanic['age'] >= 20
titanic[upper_age_mask & lower_age_mask]
```

```python
# Another way to do above operation is by using between()
youth_mask = titanic['age'].between(20, 25)
titanic[youth_mask]
```

```python
# we can exclude or include the upper and lower bound elements by using inclusive 'both' or 'neither' respectively
youth_mask = titanic['age'].between(20, 25, inclusive='neither')
titanic[youth_mask]
```

### isin(), isnull() & notnull()

```python
cols = ['name', 'survived', 'pclass', 'age', 'fare', 'sex']
titanic = pd.read_excel('titanic3.xls', usecols=cols, index_col='name')
titanic
```

```python
# Let's select the passenger who are in 1st or 2nd class
first_mask = titanic['pclass'] == 1
second_mask = titanic['pclass'] == 2
titanic[first_mask | second_mask]
```

```python
# A better way to do achieve the same result using isin()
titanic['pclass'].isin([1, 2])
```

### How to find null

```python
titanic['age'].isnull()   # Check whether a value is null
```

```python
titanic['age'].notnull()   # Check whether a value is not null
```

```python
unknown_age_mask = titanic['age'].isnull()
titanic[unknown_age_mask]
```

```python
known_age_mask = titanic['age'].notnull()
titanic[known_age_mask]
```
